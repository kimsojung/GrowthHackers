```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as op

'''
GD로 1차원 함수 최저점 찾기
'''

#1차원 함수 정의
def f1(x):
    return (x-2)**2 + 2

#x값 할당
xx = np.linspace(-4, 4, 100)

#그래프 그리기
plt.plot(xx, f1(xx), 'k-')
plt.xlim(-5, 5)
plt.ylim(0, 40)

#도함수 정의
def df1(x):
    return 2*x - 4

#도함수 그래프 그리기
alpha = 0.4
x = 0
plt.plot(x, f1(x), 'go', markersize=10)
print("x = {}, g = {}".format(x, df1(x)))
x = x - alpha*df1(x)
plt.plot(x, f1(x), 'go', markersize=10)
print("x = {}, g = {}".format(x, df1(x)))
x = x - alpha*df1(x)
plt.plot(x, f1(x), 'go', markersize=10)
print("x = {}, g = {}".format(x, df1(x)))
plt.show()

#Scipy를 이용한 최적화
result = op.minimize(f1, 1)
print(result)
```

```python
'''
GD로 2차원 함수 최저점 찾기
'''

#2차원 함수 정의
def f2(x, y):
    return (1-x)**2 + 100*(y-x**2)**2
xx = np.linspace(-3, 3, 100)
yy = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(xx, yy)
Z = f2(X, Y)

#2차원 그래프 그리기
plt.contour(X, Y, Z, colors='gray', levels=[0.7, 3, 5, 15, 50, 150, 500, 1500, 5000])

#도함수 정의하기
def df2(x, y):
    return np.array((2*(x-1) - 400*x*(y-x**2), 200*(y-x**2)))

#도함수 그래프 그리기
alpha = 4e-4
x, y = -2, -2
g = df2(x, y)
plt.plot(x, y, 'go', markersize=10)
print("x = {}, g = {}".format(x, df1(x)))
x -= alpha*g[0]
y -= alpha*g[1]
plt.plot(x, y, 'go', markersize=10)
print("x = {}, g = {}".format(x, df1(x)))
x -= alpha*g[0]
y -= alpha*g[1]
plt.plot(x, y, 'go', markersize=10)
print("x = {}, g = {}".format(x, df1(x)))
plt.show()

#Scipy를 이용한 최적화
def f2(x):
    return (1-x[0])**2 + 100*(x[1]-x[0]**2)**2
result = op.minimize(f2, (2,2)) #두 개로 쓸거면 array로 넣어야 함
print(result)
```
